```python
import numpy as np

def ln_h1_h0(nu, sigma):
    h = 1.5 * 10**(-3)
    a = 22.5 * 10**(-3)
    delta = 1 / (sigma * 4 * 3.1415 * 10**-7* 3.1415 * nu) ** 0.5
    alpha = complex(1, 1) / delta
    return 1 / abs(np.cosh(alpha * h) + 0.5*a*alpha * np.sinh(h * alpha))

x_th = np.linspace(0.5, 30000, 1000)
# print(', '.join(map(str, x_th.ravel())))
y_th = ln_h1_h0(x_th, 4.87*10**7)
print(', '.join(map(str, y_th.ravel())))
```
